```python
import requests

search_url = "http://www.law.go.kr/DRF/lawSearch.do"
params = {
    "OC": 'devjh98',
    "target": "law",
    "type": "json",
    "query": '건축법'
}

response = requests.get(search_url, params=params)
with open('response.json', 'w', encoding='utf-8') as f:
    f.write(response.text)
```
